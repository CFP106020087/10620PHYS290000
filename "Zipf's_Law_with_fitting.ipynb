```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
import pandas 
words={}
fp=open('Alice.txt','r')
line=fp.readline()
while line :
    s=line.split()
    for a in s :
        if a in words:
         words[a]=words[a]+1
        else:
         words[a]=1
         
    
    line=fp.readline()
fp.close()
num=[]
for key in words :
        num.append(words[key])
        
print(len(num))
num.sort()
num.reverse()

print(num.index)
plt.xlabel('Rank of the word')
plt.ylabel('Number of the occurance')
plt.title(" Zipf'S Law ")
plt.xscale('log')
plt.yscale('log')
plt.plot(num)
plt.show()


def zipf_distribution(x,a,b) :
    return a*x**b      #第n(n rank)常見的頻率是最常見頻率(first rank)出現次數的1/n
xdata=np.linspace(1,len(num),len(num))

optimize_paramiter1,optimize_paramiter2=optimize.curve_fit(zipf_distribution,xdata,num)
print(optimize_paramiter1)
print(optimize_paramiter2)

fitted_curve=zipf_distribution(xdata,optimize_paramiter1[0],optimize_paramiter1[1])
plt.loglog(fitted_curve)
plt.loglog(num)
plt.xlabel('Rank of the word')
plt.ylabel('Number of the occurance')
plt.title(" Zipf'S Law ")
plt.show()



```
