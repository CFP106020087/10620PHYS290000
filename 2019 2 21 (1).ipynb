```python
import numpy as np
import matplotlib.pyplot as plt
import math
h=6.62*10**-34 #SI unit 
c=3*10**8
k=1.38*10**-23
v=np.linspace(10**0,10**15,1000,endpoint=True)
T=5000
B=2*h*v**3/c**2/(math.e**(h*v/k/T)-1)
plt.xlim(10**14,10**15)
plt.plot(v,B)
plt.xlabel("Hz")
plt.ylabel("watts")

dB=np.random.normal(0.,10**-8.5,1000)
B += dB
plt.plot(v,B)
plt.show()
```
