```python
import matplotlib.pyplot as plt
words={}
fp=open('Alice.txt','r')
line=fp.readline()
while line :
    s=line.split()
    for a in s :
        if a in words:
         words[a]=words[a]+1
        else:
         words[a]=1
    
    line=fp.readline()
fp.close()
num=[]
for key in words :
        num.append(words[key])
        
print(len(num))
num.sort()
num.reverse()


plt.xlabel('Rank of the word')
plt.ylabel('Number of the occurance')
plt.title(" Zipf'S Law ")
plt.xscale('log')
plt.yscale('log')
plt.plot(num)
plt.show()

```
