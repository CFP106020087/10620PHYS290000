```python
import numpy as np
import matplotlib.pyplot as plt
import random as r
import sys as s

#version (a) 不可負債版本 錢不夠賭資就退出

n=int(input("number of student"))
m=float(input("salary"))
k=int(input("number of games"))
dm=float(input("budget"))
a=np.repeat(m,n)
x=list(range(0,n))
#y=r.choices(x,weights=None,cum_weights=None, k=2) 任選兩人出來比
print(str(a)+"  "+"最初的本金/賭資")
i=1    
if m<dm :
    print("沒有人有錢賭博喔 ww")
    
    s.exit()
      

while i<=k :
    y=r.choices(x,weights=None,cum_weights=None, k=2)
    if y[0]==y[1]:      #任選出來的兩人不能是同個人
        
        continue
    elif a[y[0]]<dm :   #錢不夠就不能賭了
        
        continue
    elif a[y[1]]<dm :
        
        continue
    else :
        
     g=r.choice(np.arange(1,4))
    print(g)
    
    if g==1 :    #平手機率為1/3 game tie
             a[y[0]]=a[y[0]] 
             a[y[1]]=a[y[1]]
            
             print(str(y)+"下場比賽出賽者")
             print("此場比賽出現平局")
             print(str(a)+"  "+"目前每人的賭資")
    else  :
      a[y[0]]=a[y[0]]-dm
      a[y[1]]=a[y[1]]+dm
      
      print(str(y)+"下場比賽出賽者")
      print("此場比賽沒有平手")
      print(str(a)+"  "+"目前每人的賭資") 
    i+=1
        

    
    
    
plt.xlabel("student's budget")
plt.ylabel("number of students")
plt.hist(a,bins=100)
plt.show()



```

```python
import numpy as np 
import matplotlib.pyplot as plt 
import random as r

#version(b) 可負債版本

n=int(input("number of student")) 
m=float(input("salary")) 
k=int(input("number of games")) 
dm=float(input("budget")) 
a=np.repeat(m,n) 
x=list(range(0,n)) 
#y=r.choices(x,weights=None,cum_weights=None, k=2) 

print(a) 

i=1 
while i<k :
        
    y=r.choices(x,weights=None,cum_weights=None, k=2) 
    if y[0]==y[1]: #自己跟自己玩不了遊戲
        
     continue
    
    elif r.choice(np.arange(1,4))==1: #平手機率1/3
        a[y[0]]=a[y[0]]
        a[y[1]]=a[y[1]]
        print(str(y)+"下場比賽出賽者")
        print("此場比賽出現平局")
        print(str(a)+"  "+"目前每人的賭資")
        
    else :
        a[y[0]]=a[y[0]]-dm 
        a[y[1]]=a[y[1]]+dm 
        
        print(str(y)+"下場比賽出賽者")
        print("此場比賽沒有平手")
        print(str(a)+"  "+"目前每人的賭資")
    
    i=i+1

plt.xlabel("student's budget") 
plt.ylabel("number of students") 
plt.hist(a,bins=100) 
plt.show()
```
